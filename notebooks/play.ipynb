```python
from pathlib import Path
import csv
import matplotlib.pyplot as plt
```

```python
train_logfile_path = Path().absolute().parent / "logs" / "run_20250314_gftbpc_train.csv"
valid_logfile_path = Path().absolute().parent / "logs" / "run_20250314_gftbpc_valid.csv"

train_steps, train_loss = [], []
f = open(train_logfile_path)
r = csv.reader(f)
headers = next(r)
for i, line in enumerate(r): 
    train_steps.append(int(line[0]))
    train_loss.append(float(line[2]))

valid_steps, valid_loss = [], []
f = open(valid_logfile_path)
r = csv.reader(f)
headers = next(r)
for line in r:
    valid_steps.append(int(line[0]))
    valid_loss.append(float(line[1]))
    
train_start = 2000
valid_start = 8 
plt.plot(train_steps[train_start:], train_loss[train_start:])
plt.plot(valid_steps[valid_start:], valid_loss[valid_start:])
```
